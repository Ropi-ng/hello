```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

## 1. Time Series Prep

```python
# Load risk free rate data
eff_rate = pd.read_excel('effr.xlsx',index_col=None)
```

```python
# Choose the date and rate column
eff_rate = eff_rate[['Effective Date','Rate (%)']]
```

```python
# Turn year rate into daily rate
eff_rate['Rate (%)'] = eff_rate['Rate (%)'].apply(lambda x: x/252)
```

```python
# Reindex
eff_rate = eff_rate.set_index('Effective Date')
```

```python
eff_rate.index.names = ['Date']
eff_rate = eff_rate.iloc[::-1]
```

```python
eff_rate.index = pd.to_datetime(eff_rate.index)
```

```python
eff_rate.plot()
plt.xticks(rotation = 45)
plt.xlabel('Date')
plt.ylabel('Rate %')
plt.title('Effective Fed Funds Rate',fontsize='large',fontweight='bold')
plt.show()
```

```python
# Download S&P 500 ETF
snp500_ETF = yf.download(tickers='SPY',start='2014-01-01',end='2019-12-31')
```

```python
# Choose the close value
snp500_ETF = snp500_ETF['Adj Close']
```

```python
snp500_ETF = snp500_ETF.to_frame()
```

```python
assets = pd.merge(snp500_ETF,eff_rate,how='left',left_index=True,right_index=True)
assets = assets.fillna(method='ffill')
```

```python
assets['Return'] = snp500_ETF.pct_change()
```

```python
snp500_ETF_return = assets['Return']
```

```python
snp500_ETF_return = snp500_ETF_return.to_frame()
```

```python
snp500_ETF_return.plot()

plt.show()
```

```python
assets['Excess return'] = snp500_ETF_return['Return'].sub(eff_rate['Rate (%)'].apply(lambda x: x/100),axis=0)
```

```python
snp500_ETF_excessreturn = assets['Excess return'].dropna()
```

```python
snp500_ETF_excessreturn.plot()
plt.xlabel('Date')
plt.ylabel('Excess return')
plt.title('Excess return of S&P 500 ETF')
plt.show()
```

## 2. Trading Strategies

Bollinger bands

```python
#Split samples into training and test set by 7:3
training_set = assets.iloc[:1056,:]
test_set = assets.iloc[1056:,:]
```

```python
# Retrieve the data for the SPDR
data = training_set.copy()

# Compute the 20-day moving average
data['MA20'] = data['Adj Close'].rolling(window=20).mean()

# Compute the 20-day standard deviation
data['STD20'] = data['Adj Close'].rolling(window=20).std()

# Compute the upper and lower Bollinger Bands
data['UpperBand'] = data['MA20'] + 2 * data['STD20']
data['LowerBand'] = data['MA20'] - 2 * data['STD20']
```

```python
# Initialize the portfolio value and the position
V0 = 200000
portfolio_value = V0
position = 0

# Create lists to store the values of the portfolio and the position over time
portfolio_values = [portfolio_value]
positions = [position]

# Iterate over the data
for i in range(20, len(data)):
    # Buy when the price is below the lower Bollinger Band
    if data['Adj Close'][i] < data['LowerBand'][i]:
        # Compute the number of shares to buy
        shares = int(portfolio_value * L / data['Adj Close'][i])
        # Buy the shares and update the position and portfolio value
        position += shares
        portfolio_value -= shares * data['Adj Close'][i]
    # Sell when the price is above the upper Bollinger Band
    elif data['Adj Close'][i] > data['UpperBand'][i]:
        # Sell all the shares and update the position and portfolio value
        portfolio_value += position * data['Adj Close'][i]
        position = 0
    # Compute the value of the position and the portfolio at the end of the day
    position_value = position * data['Adj Close'][i]
    portfolio_value = position_value + (V0 - position_value) * data['Adj Close'][i] / data['Adj Close'][20]
    # Store the values of the portfolio and the position
    portfolio_values.append(portfolio_value)
    positions.append(position)
```

```python
#We take 20 days time window
time_window = 20
cumsum = [0]
"""
def bollinger_bands(df):
    
    data=copy.deepcopy(df)
    data['std']=data['price'].rolling(window=20,min_periods=20).std()
    data['mid band']=data['price'].rolling(window=20,min_periods=20).mean()
    data['upper band']=data['mid band']+2*data['std']
    data['lower band']=data['mid band']-2*data['std']
    
    return data
"""
ma = np.zeros(len(training_set))
upper_bound = np.zeros(len(training_set))
lower_bound = np.zeros(len(training_set))

w = np.zeros(len(training_set))
cash = np.zeros(len(training_set)+1)
theta = np.zeros(len(training_set)+1)

cash[0] = 200000
L = 5
theta[0] = 0

for i,x in enumerate(training_set['Adj Close'], 0):
    cumsum.append(cumsum[i] + x)
    ma[i] = x
    if i>=time_window:
        moving_ave = (cumsum[i] - cumsum[i-time_window])/(time_window)
        ma[i] = moving_ave
        std = training_set['Adj Close'].iloc[i-time_window:i].std()
        upper_bound[i] = ma[i] + 2 * std
        lower_bound[i] = ma[i] - 2 * std
        if w[i] * x/theta[i] > 0.82 or theta[i] == 0:
            if upper_bound[i] < x and theta[i] >= 0:
                theta[i+1] = - cash[i] * L
                w[i+1] = - cash[i] * L/x
                cash[i+1] = cash[i]  + theta[i+1] - theta[i]

            elif lower_bound[i] > x and theta[i]<=0:
                theta[i+1] = cash[i] * L
                cash[i+1] = cash[i] + theta[i+1] - theta[i]
                w[i+1] = cash[i] * L/x
            else:
                cash[i+1] = cash[i] + theta[i] * training_set['Excess return'][i]
                theta[i+1] = theta[i]
        else:
            theta[i+1] = 0
            w[i+1] = 0
    else:
        theta[i+1] = theta[i]
        cash[i+1] = cash[i]
    
```
